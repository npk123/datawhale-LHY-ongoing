```python
import numpy as np
import pandas as pd
```

```python
rd=pd.read_csv('train.csv')
```

```python
rd
```

```python
rd1=rd[rd['observation']=='PM2.5']
```

```python
rd1
```

```python
rd2=rd1.iloc[:,3:]
rd2
```

```python
train_x=[]
train_y=[]
for i in range(15):
    x=rd2.iloc[:,i:i+9]
    x.columns=np.array(range(9))  
    y=rd2.iloc[:,i+9]
    y.columns=np.array(range(1))
    train_x.append(x)
    train_y.append(y)
```

```python
train_x=pd.concat(train_x)
train_y=pd.concat(train_y)  #train_x and train_y are the type of Dataframe
train_x
```

```python
train_x=np.array(train_x,float)
train_y=np.array(train_y,float)
train_x
```

```python
#add bias to train dataset
train_x=np.concatenate((np.ones((train_x.shape[0],1)),train_x),axis=1)
train_x
```

```python
w=np.zeros(len(train_x[0]))
lr=10
Iteration=10000
sum_gra=np.zeros(len(train_x[0]))  #define sum=0
for i in range(Iteration):
    y_new=np.dot(train_x,w)
    loss=y_new-train_y
    gra= 2*np.dot(train_x.transpose(),loss) #notice we must transpose train_x 
    sum_gra += gra**2
    ada=np.sqrt(sum_gra)
    w=w-lr*gra/ada
w    
```

```python
rd_test=pd.read_csv('test.csv')
rd_test.head()
```

```python
rd2=rd_test[rd_test['AMB_TEMP']=='PM2.5']
rd2
```

```python
PMdata_test=rd2.iloc[:,2:]
PMdata_test
```

```python
test_x=np.array(PMdata_test,float)
test_x
```

```python
test_x=np.concatenate((np.ones((test_x.shape[0],1)),test_x),axis=1)
test_x
```

```python
y_pre=np.dot(test_x,w)
y_pre
```

```python
y_submit=pd.read_csv('sampleSubmission.csv')
y_submit.value=y_pre   #get value by .value   automatically identify numbers
y_submit.to_csv('sampleSubmission.csv',index=False)
```

```python
real=pd.read_csv('answer.csv')
error=abs(y_submit.value-real.value).sum()/len(real.value)
error
```
